Title: Implementation of Feed Forward Neural Network using SLP and MLP

Aim:
To implement and understand Feed Forward Neural Networks using Single Layer Perceptron (SLP) and Multi-Layer Perceptron (MLP).

Procedure:

1.Import Required Libraries
Import TensorFlow/Keras or NumPy.

2.Prepare Dataset
Load a simple dataset (e.g., AND/OR logic or MNIST dataset).

3.Implement SLP (Single Layer Perceptron)
Create a model with only one dense layer.
Use an activation function like sigmoid or softmax.

4.Compile the Model
Define optimizer, loss function, and metrics.

5.Train the Model
Train using training data.

6.Evaluate the Model
Test the model using test data.

7.Implement MLP (Multi-Layer Perceptron)
Add one or more hidden layers.
Use activation functions like ReLU.

8.Compare Results
Observe performance difference between SLP and MLP.

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Sample dataset (AND logic)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 0, 0, 1])

# -------- SLP Model --------
slp_model = Sequential([
    Dense(1, activation='sigmoid', input_shape=(2,))
])

slp_model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

slp_model.fit(X, y, epochs=100, verbose=0)

print("SLP Predictions:")
print(slp_model.predict(X))

# -------- MLP Model --------
mlp_model = Sequential([
    Dense(4, activation='relu', input_shape=(2,)),
    Dense(1, activation='sigmoid')
])

mlp_model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

mlp_model.fit(X, y, epochs=100, verbose=0)

print("\nMLP Predictions:")
print(mlp_model.predict(X))
```
