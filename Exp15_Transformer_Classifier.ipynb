```python
Experiment No: 15

Title: Transformer-Based Classifier

Aim:

To implement a Transformer-based classifier for text classification and understand how attention mechanisms improve model performance.

Procedure:

1.Import Libraries
Import TensorFlow/Keras and required modules.

2.Load Dataset
Use a text dataset (e.g., IMDB sentiment dataset).

3.Preprocess Data
Tokenize text and convert into sequences.

4.Build Transformer Model
Use embedding layer
Apply Multi-Head Attention
Add feed-forward layers

5.Add Classification Layer
Use Dense layer with softmax/sigmoid activation.

6.Compile Model
Define optimizer, loss, and metrics.

7.Train Model
Train using training dataset.

8.Evaluate Model
Test model accuracy on test dataset.
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.datasets import imdb

# Load dataset
vocab_size = 10000
max_len = 200

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)

# Pad sequences
x_train = tf.keras.preprocessing.sequence.pad_sequences(x_train, maxlen=max_len)
x_test = tf.keras.preprocessing.sequence.pad_sequences(x_test, maxlen=max_len)

# Transformer block
class TransformerBlock(layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim):
        super().__init__()
        self.att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            layers.Dense(ff_dim, activation='relu'),
            layers.Dense(embed_dim),
        ])
        self.norm1 = layers.LayerNormalization()
        self.norm2 = layers.LayerNormalization()

    def call(self, inputs):
        attn_output = self.att(inputs, inputs)
        out1 = self.norm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        return self.norm2(out1 + ffn_output)

# Model
embed_dim = 32
num_heads = 2
ff_dim = 32

inputs = layers.Input(shape=(max_len,))
x = layers.Embedding(vocab_size, embed_dim)(inputs)
x = TransformerBlock(embed_dim, num_heads, ff_dim)(x)
x = layers.GlobalAveragePooling1D()(x)
x = layers.Dense(20, activation="relu")(x)
outputs = layers.Dense(1, activation="sigmoid")(x)

model = tf.keras.Model(inputs, outputs)

# Compile
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"])

# Train
model.fit(x_train, y_train, epochs=3, batch_size=32)

# Evaluate
loss, acc = model.evaluate(x_test, y_test)
print("Accuracy:", acc)
```
