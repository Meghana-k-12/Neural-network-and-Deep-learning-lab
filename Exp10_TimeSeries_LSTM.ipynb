Title: Time Series Forecasting using LSTM

Aim:

To implement a Long Short-Term Memory (LSTM) model for time series forecasting and predict future values based on past sequential data.

Procedure:

1.Import Libraries
Import TensorFlow/Keras, NumPy, and Matplotlib.

2.Prepare Dataset
Create or load time series data (e.g., stock prices or sequence data).

3.Preprocess Data
Normalize data and convert into sequences (input-output pairs).

4.Reshape Data
Convert data into 3D format (samples, timesteps, features).

5.Build LSTM Model
Create a Sequential model with LSTM layer(s).

6.Add Dense Layer
Add output layer for prediction.

7.Compile Model
Use optimizer (Adam) and loss function (MSE).

8.Train Model
Train the model using training data.

9.Predict Values
Generate predictions for future values.

10.Visualize Results
Plot actual vs predicted values.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Generate sample data
data = np.arange(1, 51)

# Prepare dataset
X, y = [], []
for i in range(len(data)-3):
    X.append(data[i:i+3])
    y.append(data[i+3])

X = np.array(X)
y = np.array(y)

# Reshape to [samples, timesteps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))

# Build LSTM model
model = Sequential([
    LSTM(50, activation='relu', input_shape=(3,1)),
    Dense(1)
])

# Compile model
model.compile(optimizer='adam', loss='mse')

# Train model
model.fit(X, y, epochs=200, verbose=0)

# Predict
predicted = model.predict(X)

# Plot results
plt.plot(y, label='Actual')
plt.plot(predicted, label='Predicted')
plt.legend()
plt.show()
```

Result:

Successfully implemented time series forecasting using an LSTM model. The model effectively captured long-term dependencies and predicted future values with improved accuracy compared to basic RNN.
