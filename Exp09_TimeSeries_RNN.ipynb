Title: Time Series Forecasting using RNN

Aim:

To implement a Recurrent Neural Network (RNN) model for time series forecasting and predict future values based on past sequential data.

Procedure:

1.Import Libraries
Import TensorFlow/Keras, NumPy, and Matplotlib.

2.Prepare Dataset
Create or load time series data (e.g., stock prices or sequence data).
3.Preprocess Data
Normalize data and convert it into sequences (input-output pairs).

4.Reshape Data
Convert data into 3D format (samples, timesteps, features).

5.Build RNN Model
Create a Sequential model with SimpleRNN layer.

6.Add Dense Layer
Add output layer for prediction.

7.Compile Model
Define optimizer (Adam) and loss function (MSE).

8.Train Model
Train using training data.

9.Predict Future Values
Use trained model to predict next values.

10.Visualize Results
Plot actual vs predicted values.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# Generate sample time series data
data = np.arange(1, 51)

# Prepare dataset
X = []
y = []

for i in range(len(data)-3):
    X.append(data[i:i+3])
    y.append(data[i+3])

X = np.array(X)
y = np.array(y)

# Reshape for RNN [samples, timesteps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))

# Build model
model = Sequential([
    SimpleRNN(10, activation='relu', input_shape=(3,1)),
    Dense(1)
])

# Compile
model.compile(optimizer='adam', loss='mse')

# Train
model.fit(X, y, epochs=200, verbose=0)

# Predict
predicted = model.predict(X)

# Plot
plt.plot(y, label='Actual')
plt.plot(predicted, label='Predicted')
plt.legend()
plt.show()
```

Result:

Successfully implemented time series forecasting using an RNN model. The model effectively learned sequential patterns and predicted future values with good accuracy.
