Title: Implement the RNN Deep Learning Model

Aim:

To implement a Recurrent Neural Network (RNN) model and understand its ability to process sequential data.

Procedure:

1.Import Libraries
Import TensorFlow/Keras and required modules.

2.Prepare Dataset
Use a simple sequential dataset (e.g., text or time-series data).

3.Preprocess Data
Convert data into sequences and normalize if required.

4.Build RNN Model
Create a Sequential model and add SimpleRNN layers.

5.Add Dense Layer
Add an output layer for prediction.

6.Compile Model
Define optimizer, loss function, and metrics.

7.Train Model
Train the model using training data.

8.Evaluate Model
Test the model and observe predictions.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# Sample sequential data
X = np.array([
    [[1], [2], [3]],
    [[2], [3], [4]],
    [[3], [4], [5]],
    [[4], [5], [6]]
])

y = np.array([4, 5, 6, 7])

# Build RNN model
model = Sequential([
    SimpleRNN(10, activation='relu', input_shape=(3, 1)),
    Dense(1)
])

# Compile model
model.compile(optimizer='adam', loss='mse')

# Train model
model.fit(X, y, epochs=200, verbose=0)

# Test prediction
test_input = np.array([[[5], [6], [7]]])
prediction = model.predict(test_input)

print("Predicted value:", prediction)
```

Result:

Successfully implemented a Recurrent Neural Network (RNN) model. The model learned sequential patterns and was able to predict the next value in a sequence effectively.
