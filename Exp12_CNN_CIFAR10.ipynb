Title: CNN with CIFAR Dataset

Aim:

To implement a Convolutional Neural Network (CNN) using the CIFAR-10 dataset for image classification.

Procedure:
1.Import Libraries
Import TensorFlow, Keras, NumPy, and Matplotlib.

2.Load Dataset
Load the CIFAR-10 dataset containing 60,000 images of 10 classes.

3.Preprocess Data
Normalize pixel values (0вЂ“255 в†’ 0вЂ“1).

4.Define Class Labels
Assign labels like airplane, car, dog, etc.

5.Build CNN Model
Add convolutional layers, pooling layers, flatten, and dense layers.

6.Compile Model
Define optimizer (Adam), loss function, and accuracy metric.

7.Train Model
Train the model using training data.

8.Evaluate Model
Test model performance on test dataset.

9.Predict Output
Predict classes for sample images.

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()

# Normalize
x_train, x_test = x_train / 255.0, x_test / 255.0

# Class names
class_names = ['airplane','automobile','bird','cat','deer',
               'dog','frog','horse','ship','truck']

# Build CNN model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
history = model.fit(x_train, y_train, epochs=5,
                    validation_data=(x_test, y_test))

# Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test)

print("Test Accuracy:", test_acc)

# Predict sample
pred = model.predict(x_test[:1])
print("Predicted class:", class_names[np.argmax(pred)])
```

Result:

Successfully implemented a CNN model using the CIFAR-10 dataset. The model effectively classified images into different categories with good accuracy, demonstrating the capability of CNNs in image classification tasks.
