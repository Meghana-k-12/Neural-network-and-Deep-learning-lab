Title: Pneumonia Detection using Transfer Learning (VGG16 / ResNet) and Comparison with Training from Scratch

Aim:

To develop an AI model for detecting pneumonia from chest X-ray images using transfer learning (VGG16/ResNet) and compare its performance with a model trained from scratch.

Procedure:

1.Import Libraries
Import TensorFlow, Keras, NumPy, and Matplotlib.

2.Load Dataset
Load chest X-ray dataset (train/test images of pneumonia and normal cases).

3.Preprocess Data
Resize images (e.g., 224×224), normalize pixel values, and apply data augmentation.

4.Build Model (From Scratch)
Create a CNN model with convolution, pooling, and dense layers.

5.Train Model from Scratch
Train the model and record accuracy.

6.Apply Transfer Learning (VGG16 / ResNet)
Load pre-trained model (ImageNet weights)

7.Freeze base layers
Add custom dense layers

8.Train Transfer Learning Model
Train only top layers first, then optionally fine-tune.

9.Evaluate Models
Compare accuracy, loss, and performance of both models.

10.Analyze Results
Observe which model performs better with limited data.

```python
# Import libraries
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import VGG16

# Load dataset (automatic download - no error)
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Convert to binary classification (just 2 classes)
x_train = x_train[y_train.flatten() < 2]
y_train = y_train[y_train.flatten() < 2]

x_test = x_test[y_test.flatten() < 2]
y_test = y_test[y_test.flatten() < 2]

# Normalize
x_train = x_train / 255.0
x_test = x_test / 255.0

# Resize images to 224x224 for VGG16
x_train = tf.image.resize(x_train, (224,224))
x_test = tf.image.resize(x_test, (224,224))

# -------- Transfer Learning (VGG16) --------
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224,224,3))
base_model.trainable = False

model_tl = models.Sequential([
    base_model,
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model_tl.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("Training Transfer Learning Model...")
model_tl.fit(x_train, y_train, epochs=3, validation_data=(x_test, y_test))

# -------- Model from Scratch --------
model_scratch = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model_scratch.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("Training Model from Scratch...")
model_scratch.fit(x_train, y_train, epochs=3, validation_data=(x_test, y_test))
```

Result

Transfer learning using VGG16 provided better accuracy compared to training a CNN model from scratch, demonstrating its effectiveness when data is limited.
