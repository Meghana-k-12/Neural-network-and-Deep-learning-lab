Title: Implementation of Simple Vector Manipulation

Aim:
To implement and understand basic vector operations such as addition, subtraction, multiplication, and dot product using Python.

Procedure:

1.Import Library
Import NumPy or TensorFlow for vector operations.

2.Create Vectors
Define two vectors with numerical values.

3.Perform Vector Addition
Add corresponding elements of the vectors.

4.Perform Vector Subtraction
Subtract elements of one vector from another.

5.Perform Element-wise Multiplication
Multiply corresponding elements.

6.Find Dot Product
Calculate the dot product of two vectors.

7.Display Results
Print all outputs for verification.

```python
import numpy as np

# Create vectors
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# Vector operations
addition = v1 + v2
subtraction = v1 - v2
multiplication = v1 * v2
dot_product = np.dot(v1, v2)

# Display results
print("Vector 1:", v1)
print("Vector 2:", v2)
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Dot Product:", dot_product)
```

Result:

Successfully implemented simple vector manipulation operations. The experiment demonstrates how vectors can be manipulated using basic mathematical operations in Python.
