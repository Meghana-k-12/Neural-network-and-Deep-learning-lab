Title: Study of Deep Learning Life Cycle

**Aim**:
To understand the different stages involved in the deep learning life cycle and how a deep learning model is developed from data collection to deployment.

**Procedure:**
1.Data Collection
Gather relevant data from various sources like datasets, sensors, or online repositories.

2.Data Preprocessing
Clean the data by handling missing values, removing noise, and normalizing or scaling features.

3.Data Splitting
Divide the dataset into training, validation, and testing sets.

4.Model Selection
Choose an appropriate deep learning model (e.g., CNN, RNN, ANN) based on the problem.

5.Model Training
Train the model using training data by adjusting weights through optimization algorithms.

6.Model Evaluation
Evaluate the model performance using metrics like accuracy, precision, recall, etc.

7.Hyperparameter Tuning
Improve model performance by adjusting parameters like learning rate, batch size, etc.

8.Deployment
Deploy the trained model into real-world applications.

9.Monitoring & Maintenance
Continuously monitor the model and update it with new data if required.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load dataset (MNIST)
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Build model
model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train model
model.fit(x_train, y_train, epochs=5)

# Evaluate model
test_loss, test_acc = model.evaluate(x_test, y_test)

print("Test Accuracy:", test_acc)
```

Result:

Successfully studied the deep learning life cycle and implemented a basic deep learning model. The experiment demonstrates how data is processed and used to train, evaluate, and deploy a model.
